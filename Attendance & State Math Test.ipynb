## Is there a relationship between attendance and NY State Math Test?

```python
import pandas as pd
import datetime
```

```python
attendance = "https://raw.githubusercontent.com/aalilyah/DS4a/Admissions/admissions_2013-2022"
att = pd.read_csv(attendance, index_col = 0)

math_results = "https://raw.githubusercontent.com/aalilyah/DS4a/main/Math_Results_2015-2022.csv"
results = pd.read_csv(math_results, index_col = 0)
```

```python
results.index.name = 'DBN'
```

```python
fix_year_results = results.copy()
fix_year_results.head()

```

```python
a = fix_year_results['Year'].astype(str)
clean_year = []

for x in a:
    list1 = x[2:4]
    clean_year.append(list1)
    
fix_year_results['Year'] = clean_year
```

```python
fix_year_att = att.copy()
fix_year_att
```

```python
b = fix_year_att['Year'].astype(str)
att_clean_year = []

for x in b:
    list1 = x[5:8]
    att_clean_year.append(list1)
    
    
fix_year_att['Grade'] = fix_year_att['Grade'].fillna(0).astype(int)
fix_year_att.index = fix_year_att.index.astype(str)

fix_year_att['Year'] = att_clean_year

```

```python
fix_year_att = fix_year_att.reset_index()
fix_year_results = fix_year_results.reset_index()
```

```python
fix_year_att.DBN = fix_year_att.DBN.astype(str)
fix_year_results.DBN = fix_year_results.DBN.astype(str)

fix_year_att.Year = fix_year_att.Year.astype(str)
fix_year_results.Year = fix_year_results.Year.astype(str)

fix_year_att.Grade = fix_year_att.Grade.astype(str)
fix_year_results.Grade = fix_year_results.Grade.astype(str)

att_results = fix_year_att.merge(fix_year_results, how = 'left', on = ['DBN', 'Year', 'Grade'])
```

```python
att_results
```

```python
attvsresults = att_results.dropna(axis = 0)
```
