## Is there a relationship between attendance and NY SHSAT?

```python
import pandas as pd
import datetime
```

```python
attendance = "https://raw.githubusercontent.com/aalilyah/DS4a/main/attendance_8thgrade"
att = pd.read_csv(attendance, index_col = 0)

shsat_results = "https://raw.githubusercontent.com/aalilyah/DS4a/main/2015-2021_SHSAT_Admissions_Test_Offers_By_Sending_School.csv"
shsat = pd.read_csv(shsat_results, index_col = 0)
```

```python
att.set_index('DBN', inplace = True)
```

```python
att = att[att.index.str.contains('02M') | att.index.str.contains('24Q')  | 
                att.index.str.contains('31R') | att.index.str.contains('20K')  | 
                att.index.str.contains('10X')]
```

```python
att.drop(columns = {'Grade'} , inplace = True)
```

```python
att.to_csv('attendance_8thgrade')
```
