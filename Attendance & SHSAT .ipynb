## Is there a relationship between attendance and NY SHSAT?

```python
import pandas as pd
import datetime
```

```python
attendance = "https://raw.githubusercontent.com/aalilyah/DS4a/Admissions/admissions_2013-2022"
att = pd.read_csv(attendance, index_col = 0)

shsat_results = "https://raw.githubusercontent.com/aalilyah/DS4a/main/2015-2021_SHSAT_Admissions_Test_Offers_By_Sending_School.csv"
shsat = pd.read_csv(shsat_results, index_col = 0)
```

```python
a = shsat['Year'].astype(str)
clean_year = []

for x in a:
    list1 = x[7:9]
    clean_year.append(list1)
    
shsat['Year'] = clean_year
shsat
```

```python
att['Grade'] = att['Grade'].fillna(0).astype(int)
```

```python
b = att['Year'].astype(str)
att_clean_year = []

for x in b:
    list1 = x[5:8]
    att_clean_year.append(list1)
    
att['Year'] = att_clean_year
att
```

```python
_att = att.reset_index()
shsat_results = shsat.reset_index()
```

```python
shsat_results.rename(columns = {'School_DBN': 'DBN'}, inplace = True)
```

```python
_att.DBN = _att.DBN.astype(str)
shsat_results.DBN = shsat_results.DBN.astype(str)

_att.Year = _att.Year.astype(str)
shsat_results.Year = shsat_results.Year.astype(str)


sa = _att.merge(shsat_results, how = 'left', on = ['DBN', 'Year'])
```

```python
SHSATvsATT = sa.dropna(axis = 0)
```

```python
SHSATvsATT
```

```python
SHSATvsATT.to_csv('attendace_corr_shsat')
```
