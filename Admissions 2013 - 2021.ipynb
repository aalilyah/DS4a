## 2013-2019 and 2018-2022 Attendance

```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', 600)
```

```python
# attendance is documented yearly
url1 = "https://raw.githubusercontent.com/itsyanlee/DS4A_data/admissions/2013-2019_Attendance_Results_-_School.csv"
admis_2013_2019 = pd.read_csv(url1, index_col = 0)


# attendance is documented daily and needs to be aggregated
url2 = "https://raw.githubusercontent.com/itsyanlee/DS4A_data/admissions/2018-2021_Daily_Attendance_by_School.csv"
admis_2018_2021 = pd.read_csv(url2, index_col = 0)
```

DBN Order : K=Brooklyn, M=Manhattan, R=Staten Island, Q=Queens, X=Bronx

### Schools only in these districts 

District 31 - staten island 

District 2 - Manhattan

District 24 - Queens

District 10 - Bronx 

District 20 - Brooklyn

```python
# Schools from admis_2013_2019 

attendance_by_district_x = admis_2013_2019[admis_2013_2019.index.str.contains('02M') | admis_2013_2019.index.str.contains('24Q')  | 
                admis_2013_2019.index.str.contains('31R') | admis_2013_2019.index.str.contains('20K')  | 
                admis_2013_2019.index.str.contains('10X')]

attendance_by_district_x = attendance_by_district_x.loc[(attendance_by_district_x.Grade == '6') | 
                           (attendance_by_district_x.Grade == '7') | 
                           (attendance_by_district_x.Grade == '8')]

attendance_by_district_x = attendance_by_district_x[attendance_by_district_x['% Attendance'] != 's']
```

```python
attendance_by_district_x
```

```python
attendance_by_district_x = attendance_by_district_x[(attendance_by_district_x['Year'] != '2019-2020') | 
                        ( attendance_by_district_x['Year'] != '2020-2021' )]
```

```python
# removing columns
attendance_by_district_x.drop(columns = {'Demographic Category','Demographic Variable',
                                         '# Total Days','# Days Absent','# Days Present', 
                                         '# Contributing 20+ Total Days', '# Chronically Absent', 
                                         '% Chronically Absent', 'School Name'}, inplace = True)
```

```python
# Schools from admis_2018_2021 

attendance_by_district_y = admis_2018_2021[admis_2018_2021.index.str.contains('02M') | admis_2018_2021.index.str.contains('24Q')  | 
                admis_2018_2021.index.str.contains('31R') | admis_2018_2021.index.str.contains('20K')  | 
                admis_2018_2021.index.str.contains('10X')]

attendance_by_district_y = attendance_by_district_y[(attendance_by_district_y['Date'].str.contains('2020')) |
                                                    (attendance_by_district_y['Date'].str.contains('2021'))]
```

```python
#clean School year to represent '2020-21'
att_y = attendance_by_district_y['SchoolYear'].astype(str)
clean_year = []

for x in att_y:
    list1 = x[:4]
    list2 = x[6:8]
    clean_year.append(list1+'-'+list2)
    
attendance_by_district_y['SchoolYear'] = clean_year
```

```python
# clean up attendance_by_district_y 

attendance_by_district_y.rename(columns = {'SchoolYear': 'Year'}, inplace = True)
attendance_by_district_y.index.name = 'DBN'

attendance_by_district_y_total_days = attendance_by_district_y.groupby(['DBN', 'Year'])[['Date','Year','Enrolled','Present','Absent','Released']].sum()
attendance_by_district_y_total_days['% Attendance'] = ((attendance_by_district_y_total_days.Present / attendance_by_district_y_total_days.Enrolled)*100).round(1)

attendance_by_district_y_total_days.drop(columns = {'Enrolled','Present','Absent','Released' }, inplace = True)
```

```python
attendance_by_district_y_total_days
```

```python
attendance_by_district_x['% Attendance'] = attendance_by_district_x['% Attendance'].astype(float)
```

```python
attendance_by_district_y_total_days = attendance_by_district_y_total_days.reset_index()
```

```python
attendance_by_district_x = attendance_by_district_x.reset_index()
```

```python
attendance_by_district_y_total_days
```

```python
attendance_by_district_x.drop(columns = {'Grade'} , inplace = True)
```

```python
attendance_by_district_x
```

```python
admissions = attendance_by_district_x.merge(attendance_by_district_y_total_days, how = 'outer', on = ['DBN', 'Year', '% Attendance']).drop_duplicates()
```

```python
admissions.to_csv('attendance_2013-2022')
```
